```python
import pandas as pd
import numpy as np
import math

from IPython.display import display
```

```python
df = pd.read_excel('C:/Users/milo/Google-Academic/Melbourne - MSc Business Analytics/5th Module - Analytics Applications/Marketing Analytics/Week 1/Data_Utilities.xlsx')
df = df.drop('UserID', axis=1) # Drop first column with UserID
df = df.T
df['Attribute'] = df.index
lst = ['Origin','Origin','Origin',
       'Price','Price','Price',
       'Body','Body','Body',
       'Aftertaste','Aftertaste','Aftertaste',
       'Calories','Calories','Calories',
       'Packaging','Packaging','Packaging',
       'Glass','Glass','Glass']

df['Attribute_Name'] = lst

cols = df.columns.tolist()
cols = cols[-2:] + cols[:-2]
df = df[cols]
n_respondents = len(df.columns) -2
df
```

```python

def calc_partworth(df_i):
    df_i.columns = ['Attribute', 'Attribute_Name', 'Raw_Utility']

    temp_min = df_i.groupby('Attribute_Name')['Raw_Utility'].min()
    temp_min = pd.DataFrame(temp_min)
    temp_min["Key"] = temp_min.index
    temp_min

    df_i = pd.merge(df_i, temp_min, left_on='Attribute_Name', right_on='Key', how = "left")
    df_i['Scaled Within Attribute'] = df_i['Raw_Utility_x'] - df_i['Raw_Utility_y'] 
    df_i = df_i[['Attribute','Attribute_Name', 'Raw_Utility_x', 'Scaled Within Attribute']]

    temp_max = df_i.groupby('Attribute_Name')['Scaled Within Attribute'].max()
    temp_max = pd.DataFrame(temp_max)
    temp_max["Key"] = temp_max.index

    sum_attributes = sum(temp_max['Scaled Within Attribute'])
    df_i['Scaled Across Attributes'] = (100 * df_i['Scaled Within Attribute']) /sum_attributes

    return df_i


df0 = df.iloc[:,[0,1,317]]

calc_partworth(df0)

```

```python
df_base = calc_partworth(df.iloc[:,[0,1,2]])
df_base = df_base[['Attribute','Attribute_Name']]

for i in range(n_respondents):
    df0 = calc_partworth(df.iloc[:,[0,1,i+2]])
    df0 = df0[['Scaled Across Attributes']]
    df_base = pd.concat([df_base, df0], axis=1)

    
df_base['Attribute2'] = df_base['Attribute'].replace('Regular.1', 'Regular')
df_base["Key"] = df_base["Attribute_Name"] + df_base["Attribute2"]

df_base
```

```python
lst_awareness = profiles.ix[0,].tolist()[1:]
lst_distribution = profiles.ix[1,].tolist()[1:]
lst_distribution
```

```python
products = pd.read_excel('C:/Users/milo/Google-Academic/Melbourne - MSc Business Analytics/5th Module - Analytics Applications/Marketing Analytics/Week 1/Products.xlsx', sheetname='Products')
n_products = len(products.columns) -1
display(products)

new_product = products["KirinNew"]
products = products.drop(labels='KirinNew', axis=1) # axis 1 drops columns
display(products)

def calc_util(df_base, products, n_respondents, n_products, new_product = []):
    
    if list(new_product):
        products["KirinNew"] = new_product

    aux_df = pd.DataFrame()
    data = pd.DataFrame({"ID": range(n_products)})
    aux_df.append(data)

    for i in range(n_respondents):
        lst_brand = []
        lst_utility = []
        df_i = df_base.iloc[:,[n_respondents+3,i+2]] 

        for j in range(n_products):
            products_j = products.iloc[:,[0,j+1]]
            products_j["Key"] = products_j["Attributes"] + products_j.ix[:,1]        

            df_i_j = pd.merge(products_j, df_i, left_on='Key', right_on='Key', how = "left")

            brand = list(df_i_j)[1]
            utility = sum(df_i_j["Scaled Across Attributes"])

            lst_brand.append(brand)
            lst_utility.append(utility)

        aux_df["Brand"] = lst_brand
        aux_df["Utility" + str(i+1)] = lst_utility
        
    return aux_df
```

```python
# Only Kirin
aux_kr = calc_util(df_base, pd.DataFrame(products["Attributes"]), n_respondents, 1, new_product)
aux_kr
```

```python
# All, Kirin included
aux_df = calc_util(df_base, products, n_respondents, n_products, new_product)
aux_df
```

```python

def calc_share_pref(prod_util_df, n_respondents, n_products):
    share_pref_df = pd.DataFrame()
    data = pd.DataFrame({"ID": range(n_products)})
    share_pref_df.append(data)
    share_pref_df["Brand"] = prod_util_df["Brand"]

    for i in range(n_respondents):
        lst_share = []
        for j in range(n_products):
            lst_share.append(prod_util_df.ix[j,i+1]/sum(prod_util_df.ix[:,i+1]))        
        
        share_pref_df["Share_Preference" + str(i+1)] = lst_share
        
    lst_totals = np.sum(share_pref_df,axis=1).tolist()
    share_pref_df["Total"] = lst_totals
    
    lst_mkt_share = []
    
    k = 0
    lst_totals2 = []
    for elem in lst_totals:
        lst_totals2.append(elem*lst_awareness[k]*lst_distribution[k])
        k+=1
    
    for elem in lst_totals2:
        lst_mkt_share.append(100*elem/sum(lst_totals2))
        
        
    share_pref_df["Market_Share"] = lst_mkt_share
    
    cols = share_pref_df.columns.tolist()
    cols = cols[-2:] + cols[:-2]
    share_pref_df = share_pref_df[cols]
    
    return(share_pref_df)
      
share_pref_df = calc_share_pref(aux_df, n_respondents, n_products)
share_pref_df
```

```python
def calc_logit_share_pref(prod_util_df, n_respondents, n_products):
    exp_df = pd.DataFrame()
    data = pd.DataFrame({"ID": range(n_products)})
    exp_df.append(data)
    exp_df["Brand"] = prod_util_df["Brand"]
    
    for i in range(n_respondents):
        lst_exp = []
        for j in range(n_products):
            lst_exp.append(math.exp(prod_util_df.ix[j,i+1]))
        
        exp_df["exp_util" + str(i+1)] = lst_exp
        

    
    return(calc_share_pref(exp_df, n_respondents, n_products))
    
exp_df = calc_logit_share_pref(aux_df, n_respondents, n_products)
exp_df
    
```

```python

def calc_choice_prob(prod_util_df, n_respondents, n_products):
    choice_prob_df = pd.DataFrame()
    data = pd.DataFrame({"ID": range(n_products)})
    choice_prob_df.append(data)
    choice_prob_df["Brand"] = prod_util_df["Brand"]

    for i in range(n_respondents):
        lst_choice = []
        max_util = max(prod_util_df.ix[:,i+1])
        
        for j in range(n_products):
            if prod_util_df.ix[j,i+1] == max_util:
                lst_choice.append(1) 
            else:
                lst_choice.append(0)
        
        removed_repeated_lst_choice = []
        for elem in lst_choice:
            removed_repeated_lst_choice.append(elem/sum(lst_choice))
        
        
        choice_prob_df["Choice_Probability" + str(i+1)] = removed_repeated_lst_choice
    
    lst_totals = np.sum(choice_prob_df,axis=1).tolist()
    choice_prob_df["Total"] = lst_totals
    
    lst_mkt_share = []
    
    k = 0
    lst_totals2 = []
    for elem in lst_totals:
        lst_totals2.append(elem*lst_awareness[k]*lst_distribution[k])
        k+=1      
        
    
    for elem in lst_totals2:
        lst_mkt_share.append(100*elem/sum(lst_totals2))
        
    choice_prob_df["Market_Share"] = lst_mkt_share
    
    cols = choice_prob_df.columns.tolist()
    cols = cols[-2:] + cols[:-2]
    choice_prob_df = choice_prob_df[cols]
    
    return(choice_prob_df)

            
choice_prob_df = calc_choice_prob(aux_df, n_respondents, n_products)
choice_prob_df

```
